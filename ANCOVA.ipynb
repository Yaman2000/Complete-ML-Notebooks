```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
models = ols('Sales~C(Promotion)',df).fit() #promotion is a categorical variable
oneway = sm.stats.anova_lm(models,typ=2) #lm - linearmodel for anova and ancova FOR linear Regression we have type value=1, for anova and ancova we have type value =2
print(oneway)
```

```python
#for promotion there is significant impact on sales because F value is 17.94 which is greater than 3.04 
#along with promtion we add coupon also
#now we hve 2 independent variable
models = ols('Sales~C(Promotion) + C(Coupon)',df).fit()
```

```python
twoway = sm.stats.anova_lm(models, typ=2)
```

```python
twoway
#residuals are the error rate
#we can see with the help of table that both promotion and coupons have value 52, that means they have a significant impact on sales 
```

```python
#Condition for the use of  ancova
#we hve independent variable which is continous - clietelrating
#we hve IDVs(client rating) which is both continous + categorical
models = ols('Sales~C(Promotion) + C(Coupon) + ClietelRatings',df).fit()
ancova = sm.stats.anova_lm(models,typ=2)
ancova
#see clientrating have significant impact on sales because it having a value greater than 3.04, so we reject the null hypothesis
#hve impact
```

```python
#now cal eta^2 - to see the impact on dependent variable with help of independent variable
def anova_table(oneway):
    oneway['eta_sq']=oneway[:-1]['sum_sq']/sum(oneway['sum_sq'])
    #( sumofsquare of all columns/sumofsquare)
    cols = ['sum_sq', 'df', 'F', 'PR(>F)', 'eta_sq']
    oneway = oneway[cols]
    return oneway
```

```python
anova_table(oneway) #significance impact on the sALES
```

```python
#57% promotion is affecting the sales
anova_table(ancova) #cLIENTRATING column which is (CONT + CATEGORICAL)
```
